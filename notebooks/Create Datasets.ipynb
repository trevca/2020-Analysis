```python
import pandas as pd
import numpy as np
```

```python
election = pd.read_csv("../data/Processed_Trump_Tweets_Election.csv").drop('Unnamed: 0', axis = 1)
briefing = pd.read_csv("../data/Processed_TrumpTweets_Coronavirus_briefing.csv").drop(['Unnamed: 0', 'Unnamed: 0.1', 'user_rt_id'], axis = 1)
control = pd.read_csv("../data/Global_Control_Tweets.csv")
```

```python
def randomSubset(data, subset_amount):
    subset = np.random.choice(data.shape[0], subset_amount, replace = False)
    subset.sort()
    print(subset)
    new_frame = []
    i = 0
    for index in subset:
        try:
            row = data.iloc[index]
            if float(row['id']) != float('nan'):
                new_frame.append(row)
                i += 1
                if i*100 % subset_amount == 0:
                    print(i/subset_amount)
        except:
            print("error: " + str(i))   
    return pd.DataFrame(new_frame, columns = data.columns)
```

```python
pd.DataFrame.to_csv(randomSubset(briefing, 100000), "../data/briefing_subset.csv")
```
